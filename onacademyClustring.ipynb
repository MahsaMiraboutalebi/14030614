```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('podcasts_en.xlsx' , engine='openpyxl')
print(data)
```

```python
data['categories']
```

```python
cats = data['categories'].str.split(',')
```

```python
cats[1]
```

```python
category_set = set()

for c in cats :
  for  e in c :
    category_set.add(e)
```

```python
len(category_set)
```

```python
cat_dic = {c:[]  for c in category_set}


for c in cats:
   for col in cat_dic.keys():

      if col in c :
            cat_dic[col].append(1)
      else:
            cat_dic[col].append(0)
```

```python
cat_dic
```

```python
cat_df = pd.DataFrame(cat_dic)
```

```python
cat_df
```

```python
cat_df.sum().sort_values()
```

```python
cat_df.sum().sort_values(ascending = False).head(10)
```

```python
cat_df.sum().sort_values(ascending = False).head(10)/data.shape[0] *100
```

```python
cat_df.sum().sort_values(ascending = True).head(10)/data.shape[0] *100
```

```python
#vizulaization
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)

X_2D = pca.fit_transform(cat_df)
```

```python
np.sum(pca.explained_variance_ratio_)
```

```python
plt.scatter (X_2D[: , 0]  ,  X_2D[: , 1] , alpha=0.3)
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()
```

```python
from sklearn.cluster import KMeans
from   sklearn.cluster import MeanShift
```

```python
cls = KMeans(n_clusters = 2)
y_pred = cls.fit_predict(X_2D)
```

```python
plt.scatter (X_2D[y_pred==0 , 0]  ,  X_2D[y_pred==0 , 1] ,  c = 'red' , alpha=0.3)
plt.scatter (X_2D[y_pred==1, 0]  ,  X_2D[y_pred==1, 1] , c = 'blue' ,  alpha=0.3)
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()
```

```python
cls = MeanShift(bandwidth = 0.1)
y_pred = cls.fit_predict(X_2D)

for l in np.unique(y_pred):
    print(l)
    plt.scatter (X_2D[y_pred==l , 0]  ,  X_2D[y_pred==l , 1] ,  c =str(l/np.unique(y_pred).shape[0]) , alpha=0.3);
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()
```

```python
cls = MeanShift(bandwidth=0.5)
y_pred = cls.fit_predict(X_2D)


plt.scatter (X_2D[y_pred==0 , 0]  ,  X_2D[y_pred==0 , 1] ,  c = 'blue' , alpha=0.3)
plt.scatter (X_2D[y_pred==1, 0]  ,  X_2D[y_pred==1, 1] , c = 'red' ,  alpha=0.3)
plt.scatter (X_2D[y_pred==2 , 0]  ,  X_2D[y_pred==2 , 1] ,  c = 'green' , alpha=0.3)
plt.scatter (X_2D[y_pred==3, 0]  ,  X_2D[y_pred==3, 1] , c = 'yellow' ,  alpha=0.3)
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()
```

```python
from sklearn.cluster import  DBSCAN
```

```python
cls = DBSCAN(eps= 0.1 , min_samples=5)
y_pred = cls.fit_predict(X_2D)


plt.scatter (X_2D[y_pred==0 , 0]  ,  X_2D[y_pred==0 , 1] ,  c = 'blue' , alpha=0.3)
plt.scatter (X_2D[y_pred==1, 0]  ,  X_2D[y_pred==1, 1] , c = 'red' ,  alpha=0.3)
plt.scatter (X_2D[y_pred==2 , 0]  ,  X_2D[y_pred==2 , 1] ,  c = 'green' , alpha=0.3)
plt.scatter (X_2D[y_pred==3, 0]  ,  X_2D[y_pred==3, 1] , c = 'yellow' ,  alpha=0.3)
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()
```
